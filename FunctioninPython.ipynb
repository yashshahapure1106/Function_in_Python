### Product of multiple numbers with list 

1. (Initialize Product with 1)
2. Initialize sum with 0

```python
def product_list(lst):
    """
     This function gives product of all elements 
    inside list containing numbers
    """
    p = 1
    for i in lst:
        p = p*i
    return p
```

```python
a = [3,4,5,6]
```

```python
c = product_list(a)
```

```python
c
```

```python
3*4*5*6
```

```python
product_list(a)
```

1. *args - Multiple inputs

```python
def func(*args) :
    for i in args:
        print(i)
```

```python
def product(*nums):
    """
    This function takes multiple numbers as input 
    and calcualtes the product (multiplication of numbers)
    """
    p = 1
    for i in nums:
        p = p*i
    return p
```

```python
product(2,3)
```

```python
product(5,25,22,26,)
```

```python
d = product(1,4,6,2,8,9,4)
```

### Write a function in python with arguments find out sum of numbers only from the input

```python
def sum_num(*args):
    """
    This functions takes multiple inputs of various datatypes and calculates sum of numbers only
    """
    s = 0
    for i in args :
        if type(i)==int or type(i)==float :
            s = s+i
    return s
```

```python
sum_num('Etlhive',3.4,12,15,11,False,12.4,15.2,'Yash')
```

```python
3.4+12+3.5+15
```

#### Write a function to create a user for website

```python
def create_user(**kwargs) :
    user = {'username':None,
            'email':None,
            'age':18}
    user.update(kwargs)
    return user
```

#### Function to calculate area of diiferent shapes

1. Circle - pi x r x r
2. Rectangle - w x h
3. Triangle - (1/2) x b x h

```python
from math import pi
```

```python
pi
```

```python
def area(shape, **kwargs) :
    for key, value in kwargs.items() :
        if shape =='circle' and key=='radius':
            ar = pi*(kwargs['radius']**2)
        elif shape =='Rectangle' and key in ['length', 'width']:
            ar = kwargs['length']*kwargs['width']
        elif shape =='triangle' and key in ['base','height']:
            ar = (1/2)*kwargs['base']*kwargs['height']
        else :
            print('Please provide correct shape and kwargs')
            print('shapes available : 1. circle 2. rectangle 3. triangle')
            ar = None
    return ar
```

```python
a1 = area(shape='circle', radius=8)
a1
```

```python
a2 = area(shape='Rectangle', length=5, width=4)
a2
```

```python
a3 = area(shape='rectangle', radius=3)
```

```python
a4 = area(shape='triangle', base=12, height=10)
a4
```

```python
(1/2)*12*10
```

```python
a5 = area(shape='hexagon', side=3)
```
